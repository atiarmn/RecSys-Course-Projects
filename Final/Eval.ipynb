```python
import re
from pathlib import Path
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pop_rec_scores = pd.read_parquet('/content/drive/MyDrive/Recsys_final/outs/rec-metrics-POP.parquet')
pop_rec_scores.head()
```

```python
tag_rec_scores = pd.read_parquet('/content/drive/MyDrive/Recsys_final/outs/rec-metrics-TAG.parquet')
tag_rec_scores.head()
```

```python
uknn_rec_scores = pd.read_parquet('/content/drive/MyDrive/Recsys_final/outs/rec-metrics-U-KNN.parquet')
uknn_rec_scores.head()
```

```python
dfs = [pop_rec_scores, tag_rec_scores, uknn_rec_scores]
res = pd.concat(dfs)
res
```

```python
plt.figure(figsize=(6, 6))

mean_metrics = res[['ndcg', 'precision', 'recall', 'hit']]

mean_metrics.plot(kind='bar')
plt.title('Average Recommendation Quality Metrics')
plt.ylabel('Score')
plt.xlabel('Algo')
plt.xticks(ticks=range(len(res['Algo'])), labels=res['Algo'], rotation=45)
plt.tight_layout()  # Adjust the plot to ensure everything fits without overlapping
plt.show()
```
